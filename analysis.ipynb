```python
# Verilerin Film / Dizi sayısını karşılaştırma
sns.countplot(x=netflix_data["type"])  # x parametresi kullanılarak dikey grafik oluşturulur
fig = plt.gcf()
fig.set_size_inches(5, 5)
plt.xticks(rotation=90)  # Kategori isimlerinin yan yazmasını engellemek için 90 derece döndürülür
plt.title("Tür")
plt.show()
```

```python
#Ratinge göre sıralama yapıyoruz
sns.countplot(x=netflix_data["rating"])  # x parametresi kullanılarak dikey grafik oluşturulur
plt.xticks(rotation=90, ha="right")  # Etiketleri 90 derece döndürüp hizalamak için
fig = plt.gcf()
fig.set_size_inches(13, 13)  
plt.title("Rating")
plt.show()
```

```python
#Tür ve rating e göre karşılaştırma
plt.figure(figsize=(10,8))
sns.countplot(x="rating",hue="type",data=netflix_data)
plt.title("Tür ve Reyting Oranları")
plt.show()
```

```python
labels=["Movie","TV Show"]
size = netflix_data["type"].value_counts()
colors =plt.cm.Wistia(np.linspace(0,1,2))
explode= [0,0.1]
plt.rcParams["figure.figsize"]=(9,9)
plt.pie(size,labels=labels,colors=colors,explode=explode,shadow=True,startangle=90)
plt.title("Tür Görünümü",fontsize=15)
plt.show()
```

```python
# 20 den fazla içeriği olan ülkeleri listeliyoruz
country_counts = netflix_data["country"].value_counts()
filtered_countries = country_counts[country_counts > 20].index
filtered_data = netflix_data[netflix_data["country"].isin(filtered_countries)]
sns.countplot(x=filtered_data["country"])  
plt.xticks(rotation=90, ha="right")  # Etiketleri 90 derece döndürüp hizalamak için
fig = plt.gcf()
fig.set_size_inches(13, 13)  
plt.title("20'den Fazla İçeriği Olan Ülkeler")
plt.show()

```

```python
# Filmleri ve dizileri ayır
movies = netflix_data[(netflix_data["type"] == "Movie") & (netflix_data["release_year"] >= 2010)]
tv_shows = netflix_data[(netflix_data["type"] == "TV Show") & (netflix_data["release_year"] >= 2010)]

movie_year_counts = movies["release_year"].value_counts()

tv_show_year_counts = tv_shows["release_year"].value_counts()
#plt.pie(size,labels=labels,colors=colors,explode=explode,shadow=True,startangle=90)
# Filmler için pasta grafiği
plt.figure(figsize=(14, 7))  # Grafiği ikiye bölmek için geniş bir boyut ayarlandı

plt.subplot(1, 2, 1)  # 1 satır, 2 sütunlu bir grafik alanında 1. grafik
plt.pie(movie_year_counts, labels=movie_year_counts.index)
plt.title("2010 Yılından Sonra Yapılan Filmlerin Yapım Yılları")

# Diziler için pasta grafiği
plt.subplot(1, 2, 2)  # 1 satır, 2 sütunlu bir grafik alanında 2. grafik
plt.pie(tv_show_year_counts, labels=tv_show_year_counts.index)
plt.title("2010 Yılından Sonra Yapılan Dizilerin Yapım Yılları")

plt.tight_layout()  # Grafikleri birbirine sığdırmak için
plt.show()
```

```python
#Kıtalara göre karşılaştırma yapıyoruz
# Ülke ve kıta eşleşmeleri
country_to_continent = {
    'United States': 'North America',
    'South Africa': 'Africa',
    'India': 'Asia',
    'United Kingdom': 'Europe',
    'Mexico': 'North America',
    'Turkey': 'Asia',
    'Austria': 'Europe',
    'Nigeria': 'Africa',
    'Japan': 'Asia',
    'France': 'Europe',
    'South Korea': 'Asia',
    'Canada': 'North America',
    'Hong Kong': 'Asia',
    'Italy': 'Europe',
    'Argentina': 'South America',
    'Brazil': 'South America',
    'Taiwan': 'Asia',
    'Poland': 'Europe',
    'Germany': 'Europe',
    'Thailand': 'Asia',
    'Indonesia': 'Asia',
    'Egypt': 'Africa',
    'Malaysia': 'Asia',
    'Philippines': 'Asia',
    'Singapore': 'Asia',
}

# Netflix verinizde kıta bilgisi eklemek
netflix_data['continent'] = netflix_data['country'].apply(lambda x: country_to_continent.get(x, 'Other'))

# Filmleri ve dizileri ayır
movies = netflix_data[netflix_data["type"] == "Movie"]
tv_shows = netflix_data[netflix_data["type"] == "TV Show"]

# Kıtalar bazında filmleri ve dizileri say
movie_continent_counts = movies["continent"].value_counts()
tv_show_continent_counts = tv_shows["continent"].value_counts()

# Filmler için pasta grafiği
plt.figure(figsize=(14, 7))

# Filmler
plt.subplot(1, 2, 1)
plt.pie(movie_continent_counts, labels=movie_continent_counts.index, autopct='%1.1f%%', startangle=90)
plt.title("Kıtalara Göre Filmler")

# Diziler
plt.subplot(1, 2, 2)
plt.pie(tv_show_continent_counts, labels=tv_show_continent_counts.index, autopct='%1.1f%%', startangle=90)
plt.title("Kıtalara Göre Diziler")

plt.tight_layout()  # Grafikleri düzenli hale getir
plt.show()
```
